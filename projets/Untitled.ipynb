```python
import pandas as pd
import numpy as np

# Charger le fichier Excel
# Afficher les données

# Charger le fichier Excel
data = pd.read_excel(r"C:\Users\Maryam\Documents\projets\Classeur2.xlsx", header=None)

# Configuration de l'affichage pour afficher toutes les lignes et éviter la notation scientifique
pd.options.display.float_format = '{:.2f}'.format

# Afficher les données
ardata=np.array(data)
input_data=ardata[:,0:8] #all data except the last column
output_data=ardata[:,8]
```

<span style="color:black"></span>

# Build machine learning Model

```python
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(12,input_dim=8,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

<span style="color:black"></span>

# Set The Configuration

```python
model.compile(loss='binary_crossentropy',optimizer='adam' , metrics=['accuracy'])
```

<span style="color:black"></span>

# Train The ML Model

```python
model.fit(input_data , output_data, epochs=150, batch_size=10,verbose=0)
```

<span style="color:black"></span>

# Calculate The Accuracy

```python
accuracy=model.evaluate(input_data,output_data,verbose=0)
print("Accuracy: %.2f %%" % (accuracy[1]*100))
```

<span style="color:black"></span>

# Make Predictions

```python
predictions=(model.predict(input_data)> 0.5).astype(int)
predictions
```

<span style="color:black"></span>

# Compare The Predicted Data with the Actual Data

```python
def display(num):
    if num==0:
        return "No Disease";
    else:
        return "Desease";

for i in range(10):
    print("%s => %s (expected %s)" % (input_data[i].tolist(),display(predictions[i]),display(output_data[i])))
```
